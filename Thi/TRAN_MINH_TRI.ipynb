```python
'''
Dữ liệu cung cấp lưu trữ thông tin của các công ty.
Thao tác chuẩn bị dữ liệu, sinh viên dùng công cụ MongoDBCompass.
Sinh viên tạo cơ sở dữ liệu mang tên companydb.
Tạo collection tblCompanies trong CSDL companydb.
Sử dụng file companies.json để import dữ liệu vào collection tblCompanies.
'''
```

```python
'''
Tiến hành viết code bên dưới câu hỏi và chạy kết quả để trả lời các câu hỏi bên dưới
'''
```

```python
# Câu 1: Viết code kết nối cơ sở dữ liệu companydb
#pip install pymongo

from pymongo import MongoClient as Mcl
myServer = Mcl('mongodb://localhost:27017')

myDB = myServer['companydb']

tblCompanies = myDB['tblCompanies']
```

```python
# Câu 2: Hiển thị danh sách name, homepage_url, category_code, email_address, phone_number của 10 công ty
listCmp = tblCompanies.find({},{
    "name":1,
    "homepage_url": 1,
    "category_code": 1,
    "email_address": 1,
    "phone_number": 1,
    "_id":0
}).limit(10)

for cmp in listCmp:
    print(cmp)
```

```python
'''
Câu 3: Hiển thị danh sách name, homepage_url, category_code, 
email_address, phone_number, number_of_employees của 10 công ty có số lượng nhân viên lớn hơn 50
'''

listCmp = tblCompanies.find({
    "number_of_employees": {"$gt": 50}
},{
    "name":1,
    "homepage_url": 1,
    "category_code": 1,
    "email_address": 1,
    "phone_number": 1,
    "number_of_employees": 1,
    "_id":0
}).limit(10)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 4: Hiển thị danh sách 10 công ty có được thành lập năm các năm 2000, 2002, 2004, 2006, 2008 
# và có số lượng nhân viên hơn 50

listCmp = tblCompanies.find({
    "number_of_employees": {"$gt": 50},
    "founded_year": {"$in":[2000, 2002, 2004, 2006, 2008]}
}).limit(10)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 5: Hiển thị danh sách name, category_code, number_of_employees, founded_year, phone_number 
# của 5 công ty được thành lập năm 2005 và sắp xếp giảm dần theo số lượng nhân viên

listCmp = tblCompanies.find({
    "founded_year": {"$eq": 2005}
},{
    "name":1,
    "category_code": 1,
    "founded_year": 1,
    "phone_number": 1,
    "number_of_employees": 1,
    "_id":0
}).sort("number_of_employees", -1).limit(5)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 6: Hiển thị danh sách 10 công ty mà thông tin chung (overview) có chữ technology
listCmp = tblCompanies.find({
    "overview": {"$regex": "/.*technology.*/"}
}).limit(10)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 7: Hiển thị danh sách 10 công ty mà tên (name) bắt đầu bằng chữ P và 
# sắp xếp tăng dần theo năm thành lập (founded_year)
# nhưng giảm dần theo số lượng nhân viên (number_of_employees)

# listCmp = tblCompanies.aggregate([
#     {"$sort": {"number_of_employees": -1, "founded_year" : 1}},
#     { "$limit" : 10 }
# ])
import pymongo

listCmp = tblCompanies.find({"name": {"$regex": "^P"}}, {"name": 1, "number_of_employees": 1, "founded_year": 1}).sort([("number_of_employees", pymongo.DESCENDING), ("founded_year", pymongo.ASCENDING)]).limit(10)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 8: Viết code chỉnh sửa thông tin số lượng nhân viên của công ty có mã lưu trữ là 52cdef7c4bab8bd675297d8a
# thành 50 nhân viên, sau đó xuất lại thông tin nhân viên có mã lưu trữ như trên

from bson.objectid import ObjectId

tblCompanies.update_one({"_id": ObjectId("52cdef7c4bab8bd675297d8a")},{"$set": {"number_of_employees": 50}})

emp = tblCompanies.find_one(
    {"_id":ObjectId("52cdef7c4bab8bd675297d8a")},
    {"number_of_employees":1}
)
print(emp)
```

```python
# Câu 9: Hiển thị danh sách 5 công ty có văn phòng (offices) đặt tại thành phố (city) là Seattle
listCmp = tblCompanies.find(
     {"offices":{"$elemMatch":{"city":"Seattle"}}}
).limit(5)

for cmp in listCmp:
    print(cmp)
```

```python
# Câu 10: Hiển thị danh sách 10 công ty sỡ hữu số lượng sản phẩm sản phẩm (products) hơn 5 

listCmp = tblCompanies.find(
     {"products":
        {"$exists" : True},
        "$where" : "this.products.length>5"}
).limit(5)

for cmp in listCmp:
    print(cmp)
```
